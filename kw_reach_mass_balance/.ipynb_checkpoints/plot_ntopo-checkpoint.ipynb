```python
%pylab inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import xarray as xr
import netCDF4 as nc4
import geopandas as gpd
```

```python
#Read data
data_path = './data/'
file_name = 'ntopo_gf_cameo.nc'
#file_name = 'ntopo_gf_conus.nc'
dx_rnet = xr.open_dataset('%s/%s' % (data_path, file_name))
print(dx_rnet)
```

```python
# Get lat/lon for segment end points
seg_id = dx_rnet['seg_id']
lon_start = dx_rnet['start_lon'].values
lat_start = dx_rnet['start_lat'].values
lon_end = dx_rnet['end_lon'].values
lat_end = dx_rnet['end_lat'].values
lon_mid = (lon_start+lon_end)/2
lat_mid = (lat_start+lat_end)/2
```

```python
#Read data
data_path = './data/'
q_mod_irf_nc = 'q_out_mod_irf.nc'
q_remap_nc = 'q_out_remap.nc'
ds_q_mod_irf = xr.open_dataset('%s/%s' % (data_path, q_mod_irf_nc))
ds_q_remap = xr.open_dataset('%s/%s' % (data_path, q_remap_nc))
```

```python
q_mod_irf=ds_q_mod_irf['IRFroutedRunoff']
q_irf=ds_q_remap['IRFroutedRunoff']
q_kwt=ds_q_remap['KWTroutedRunoff']
reachID=ds_q_mod_irf['reachID']
idx=np.where(reachID.values==14000645) #cameo
```

```python
fig, axes = plt.subplots(figsize=(10, 9), nrows=2, ncols=1)
q_diff = q_mod_irf - q_irf

q_irf[:,idx[0]].plot(ax=axes[0],label='old')
#q_kwt[:,idx[0]].plot(ax=axes[0])
q_mod_irf[:,idx[0]].plot(ax=axes[0],label='new')
axes[0].legend()
q_diff[:,idx[0]].plot(ax=axes[1],label='new-old')
axes[1].legend()
```
