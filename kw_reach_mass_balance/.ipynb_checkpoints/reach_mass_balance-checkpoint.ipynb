```python
%pylab inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Mass balance check
Procedure:  
1. This is per-segment analysis based on 365 days mizuRoute simulation 
2. Run mizuRoute (branch: mod_ntopo_output_option_working) with desireId=14000629.
3. Read in on-screen print(./data/out.txt) here.

```python
# some functions

def str2bool(v):
  return v.lower() in ("yes", "true", "t", "1")

def interp(T0,Q1,Q2,T1,T2):
    # Q1,Q2   flow at neighbouring times
    # T1,T2  ! neighbouring times
    # T0     ! desired time
    return Q1+((Q2-Q1)/(T2-T1))*(T0-T1)
```

```python
# Read out.txt

T_bound = {}
Qnew={}
Qwave0={}
Qwave={}
isRoute={}
Tentry={}
Texit={}
with open('./data/out.txt') as fin:
    t=0   
    for line in fin:
        varname = line.split()[0]
        varstr = line.split()[1:]
        if varname == 'T0-T1=':
            var = [float(str) for str in varstr]
            T_bound[t] = np.asarray(var)
        elif varname == 'Initial_Q_JRCH=':
            var = [float(str) for str in varstr]
            Qwave0[t] = np.asarray(var)
        elif varname == 'Q_JRCH=':
            var = [float(str) for str in varstr]
            Qwave[t] = np.asarray(var)
        elif varname == 'FROUTE=':
            var = [str2bool(str) for str in varstr]
            isRoute[t] = np.asarray(var)
        elif varname == 'TENTRY=':
            var = [float(str) for str in varstr]
            Tentry[t] = np.asarray(var)
        elif varname == 'T_EXIT=':
            var = [float(str) for str in varstr]
            Texit[t] = np.asarray(var)
        elif varname == 'QNEW(1)=':
            var = [float(str) for str in varstr]
            Qnew[t] = np.asarray(var)
            t+=1
```

```python
# Compute inflow, outflow and Storage at T0 (beginning of time step) and T1 (end of time step)

doPrint = False
plt_tindx = 100  # plot at plt_tindx time step for wave plots
num_tsteps = len(Qwave)

Qin = np.full(num_tsteps, np.nan)
Qout = np.full(num_tsteps, np.nan)
S0 = np.full(num_tsteps, np.nan)
S1 = np.full(num_tsteps, np.nan)

for t in range(num_tsteps):
    T0 = T_bound[t][0]
    T1 = T_bound[t][1]

    idx0 = np.where(Tentry[t]==T1)  # index where t_entry is at end of time step 
    idx1 = np.where(Texit[t]==T0)   # index where t_exit is at begining of time step
    if (not idx0 or not idx1):
        print('time step %d is not good - problem in code'%t)
        
    # Compute inflow 
    inflow = 0
    for tidx, (q, Ten) in enumerate(zip(Qwave[t], Tentry[t])):
        if T0 < Ten:  # t_entry is later than beginning of time step(T0)
            if Ten_pre >= T0: # if previous particle enters at or later than T0
                inflow = inflow + (q + q_pre)*(Ten-Ten_pre)/2
            else:  # other
                q_interp = interp(T0, q, q_pre, Ten, Ten_pre)
                inflow = inflow + (q + q_interp)*(Ten-T0)/2
        Ten_pre = Ten
        q_pre = q
    Qin[t]=inflow
    
    # Compute outflow 
    outflow = 0
    for tidx, (q, Tex) in enumerate(zip(Qwave[t], Texit[t])):
        if tidx == 0:
            Tex_pre = Tex
            q_pre = q
            continue
        if Tex <= T1:
            outflow = outflow + (q + q_pre)*(Tex-Tex_pre)/2
        elif Tex > T1:
            q_interp = interp(T1, q, q_pre, Tex, Tex_pre)
            outflow = outflow + (q_interp + q_pre)*(T1-Tex_pre)/2
            break
        Tex_pre = Tex
        q_pre = q
    Qout[t]=outflow
    
    # Compute Storage at the beginning of time step
    Stor0 = 0    
    for tidx, (q, Ten, Tex) in enumerate(zip(Qwave[t], Tentry[t], Texit[t])):
        if tidx == 0:
            Ten_pre = Ten
            q_pre = q
            continue
        if Ten <= T0:  # index where t_exit is at begining of time step
            Stor0 = Stor0 + (q + q_pre)*(Ten-Ten_pre)/2
            q_pre = q
            Ten_pre = Ten
        else:
            q_interp = interp(T0, q, q_pre, Ten, Ten_pre)
            Stor0 = Stor0 + (q_interp + q_pre)*(T0-Ten_pre)/2
            break  
    S0[t] = Stor0    
    
    # Compute Storage at the end of time step
    Stor1 = 0
    for tidx, (q, Ten, Tex) in enumerate(zip(Qwave[t], Tentry[t], Texit[t])):
        if Tex > T1 and Tex_pre < T1:
            q_interp = interp(T1, q, q_pre, Tex, Tex_pre)
            Stor1 = Stor1 + (q_interp + q)*(Tex-T1)/2
        elif Tex > T1 and Tex_pre >= T1:
            Stor1 = Stor1 + (q_pre + q)*(Tex-Tex_pre)/2
        q_pre = q
        Tex_pre = Tex
    S1[t] = Stor1 
    
    if doPrint:
        print('T0-T1   = %d - %d'%(T0,T1))
        print('inflow  = %.5f'%(inflow))
        print('outflow = %.5f'%(outflow)) 
        print('S0      = %.5f'%(Stor0)) 
        print('S1      = %.5f'%(Stor1)) 
        print('Inflow - Outflow = %.5f'%(inflow-outflow)) 
        print('dS               = %.5f'%(Stor1-Stor0)) 
        print('\n')

plt.figure(figsize=(9.5, 5.5))
plt.subplot(211)
plt.axvline(x=T_bound[plt_tindx][0], color='k', linestyle='--')
plt.axvline(x=T_bound[plt_tindx][1], color='k', linestyle='--')
plt.plot(Tentry[plt_tindx], Qwave[plt_tindx],'o')
plt.xlabel('T entry [sec]'), plt.ylabel('Qwave [m2/s]')
#plt.ylim(0, max(Qwave[plt_tindx])*1.2)
plt.title('%dth time step, '%plt_tindx)

plt.subplot(212)
plt.axvline(x=T_bound[plt_tindx][0], color='k', linestyle='--')
plt.axvline(x=T_bound[plt_tindx][1], color='k', linestyle='--')
plt.plot(Texit[plt_tindx], Qwave[plt_tindx],'o')
plt.xlabel('T exit [sec]'),plt.ylabel('Qwave [m2/s]')
#plt.ylim(0, max(Qwave[plt_tindx])*1.2)
plt.tight_layout()


plt.figure(figsize=(9.5,7.5))
plt.subplot(211)
plt.plot(Qin,'o')
plt.plot(Qout,'o')
plt.legend(('Inflow', 'Outflow'))
plt.xlabel('Time step [sec]'),plt.ylabel('Q per unit cross section [m2/day]')
plt.subplot(212)
plt.plot(Qin-Qout -(S1-S0),'o')
plt.ylim(-500, 500)
plt.ylabel('Error in mass balance [m2/day]')
```
